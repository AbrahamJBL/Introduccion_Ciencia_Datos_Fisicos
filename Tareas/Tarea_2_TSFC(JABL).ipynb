CLIENTNUM: Numero de cliente, identificador único del cliente titular de la cuenta.

Total_Revolving_Bal: Saldo rotativo total en la tarjeta de crédito

Avg_Open_To_Buy: Línea de crédito abierta para comprar (promedio de los últimos 12 meses).

Total_Amt_Chng_Q4_Q1: Cambio en el monto de la transacción (Q4 sobre Q1)

Total_Trans_Amt: Importe total de la transacción (últimos 12 meses)

Total_Trans_Ct: Recuento total de transacciones (últimos 12 meses)

Para el siguiente ejercicio se hará un analisis vibariado para identificar el numero de clientes que cuentan con las diferentes categorias de tarjetas disponibles, esto puede tener varios fines, uno de ellos incrementar mediante ofertas o publicidad la adquision de tarjetas de mayor rango, también puede ser una alerta del poco interés que se tienen por los servicios de tarjetas premium.**

```python
#con esta linea obtendremos el nombre de las categorias
tipo_tarjeta = data['Card_Category'].value_counts().index
#ahora obtendremos las veces que se repite cada categoria
usuarios = (data['Card_Category'].value_counts(normalize=True).values)*100

fig, ax = plt.subplots(figsize = (8,6))#definimos el tamaño en el que queremos el gráfico
_ = sns.barplot(x = tipo_tarjeta, y = usuarios)#le asignamos los valores a los ejes
_ = ax.set_title('Distribución de Tipos de Tarjetas de los usuarios') #definimos los ejes que se mostrarán para X y Y.
_ = ax.set_xlabel('Tipo de tarjeta')
_ = ax.set_ylabel('Porcentaje (%) de Usuarios')

#sirve para poner los numeros (porcentaje) encima de las barras	
for i in ax.containers:
    ax.bar_label(i)
```

Aunque puede ser algo obvio, de las 10127 cuentas que se evaluaron, la aplastante mayoría de poco más del 93% esta contenta con la tarjeta básica azul, y poco menos del 0.2% cuentan con la platino, y aunque entre las categorias plata y oro existe una diferencia de aproximadamente 4.3%, no creemos que sea una diferencia que se haga notar, por lo que no creemos necesario que valga la pena una inversión muy fuerte para ofertar las categorías oro y platino.

**5. Para el siguiente ejercicio se modifcará la tabla original, con el fin de crear una nueva celda, que nos marcará que con un APLICA o NO APLICA, sí el cliente tiene más de 3 meses sin usar su línea de crédito, y aplicá para una llamada o contacto con el fín de retomar su línea de crédito**

```python
#Con esta línea vemos el número de cuentas usuarios que llevan de 0 a 6 meses sin usar su cuenta
data.Months_Inactive_12_mon.value_counts()
```

Una vez con estos datos, definiremos una función que nos regrese una leyenda para la cuenta que aplique o no para un contacto y verificar por que no usa su cuenta:

```python
#definimos la funcion, donde x será valor de los meses entre 0 y 6, apartir de este valor con cadenas if regresamos el mensaje:
def aplicable_noaplica(x):
    if x > 3:
        return 'APLICA'
    else:
        return 'NO APLICA'
```

```python
#Una vez con la función, imprimimos los datos, con nuestra tabla ya modificada
#la tabla toma los valores originales y los ya modiicados, por la función "aplicable_noaplica"
data['aplicable_noaplica'] = data.Months_Inactive_12_mon.apply(aplicable_noaplica)
data #imprimimos la tabla
```

Ahora tenemos la tabla, ya se puede apreciar y hacer incluso una busqueda más rapida de las cuentas que aplican o no aplican para la atención necesaria:

```python
data.aplicable_noaplica.value_counts()#vemos cuantas cuentas estan "abandonadas" para buscar regresen a usar los servicios
```
